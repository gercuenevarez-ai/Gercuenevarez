```r
# Realizar un ANOVA para la variable Sepal.Length según la especie

anova_model <- aov(Sepal.Length ~ Species, data = iris)
print(anova_model)
summary(anova_model)

# Realizar el test de Tukey para identificar diferencias significativas

tukey_results <- TukeyHSD(anova_model)
print("Resultados del test de Tukey:")
print(tukey_results)

# Visualización del test de Tukey

plot(tukey_results, las = 1)
```

### **Determinar en R el valor de las siguientes ecuaciones**:


```r
#Ecuación 1

# Calcular 4.5 elevado a la sexta potencia
miembro1 <- 4.5^6
print(miembro1)

# Calcular la raíz cuadrada de 123
miembro2 <- sqrt(123)
print(miembro2)

# Restar los resultados de los pasos anteriores
diferencia <- miembro1 - miembro2
print(diferencia)

# Calcular el logaritmo natural de la diferencia
logaritmo <- log(diferencia)
print(logaritmo)

# Multiplicar por 3.5 (resultado final)
D <- 3.5 * logaritmo
print(D)
```

```r
#Ecuación 2

# Calcular b (raíz cuadrada de 9.2)
sqrt_9_2 <- sqrt(9.2)
print(sqrt_9_2)

# Multiplicar por 2.77
numerador_b <- 2.77 * sqrt_9_2
print(numerador_b)

b <- numerador_b / exp(1)
print(b)

# Calcular a (Elevar b a la potencia 3/4)
b_potencia <- b^(3/4)
print(b_potencia)

# Multiplicar por 5.2
numerador_a <- 5.2 * b_potencia
print(numerador_a)

# Calcular 6.7 multiplicado por la raíz cuadrada de 5
denominador_a <- 6.7 * sqrt(5)
print(denominador_a)

# Dividir numerador entre denominador (resultado final)
a <- numerador_a / denominador_a
print(a)
```

```r
# Ecuación 3

# Calcular 5.7 / 2.89
fraccion <- 5.7 / 2.89
print(fraccion)

# Sumar 5 al resultado
suma <- 5 + fraccion
print(suma)

# Calcular el logaritmo natural de la suma
log_suma <- log(suma)
print(log_suma)

# Calcular 3 elevado a la potencia 3/4
potencia_3 <- 3^(3/4)
print(potencia_3)

# Dividir el logaritmo entre 3^(3/4)
division <- log_suma / potencia_3
print(division)

# Calcular 3^34 - 3 * 3.2222 - pi
sección1 <- 3^34
sección2 <- 3 * 3.2222
sección3 <- pi
resta <- sección1 - sección2 - sección3
print(resta)

# Multiplicar los resultados de las secciones de división y resta (resultado final)
resultado <- division * resta
print(resultado)
```

```r
# Crear la función para calcular S^2

# Número de elementos
calcular_varianza <- function(x) {
  n <- length(x)
  print(n)

# Calcular la media
media <- mean(x)
  print(media)

# Restar la media
diferencias <- x - media
  print(diferencias)

# Elevar al cuadrado
  cuadrados <- diferencias^2
  print(cuadrados)

# Sumar los cuadrados
suma_cuadrados <- sum(cuadrados)
  print(suma_cuadrados)

# Dividir entre n-1
S2 <- suma_cuadrados / (n - 1)
  return(S2)
}

# Aplicación sustituyendo con datos (resultado final)
datos <- c(11, 13, 15, 17, 19)
varianza <- calcular_varianza(datos)
print(varianza)
```
